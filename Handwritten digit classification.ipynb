```python
import tensorflow.keras as tf
import matplotlib.pyplot as plt
import numpy as np
```

```python
# import the dataset
mnist = tf.datasets.mnist
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
# Create a neural network
# Create a model
model = tf.models.Sequential()
# Adding layers - input, hidden, and output
model.add(tf.layers.Flatten()) # input layer
model.add(tf.layers.Dense(784, activation = 'relu')) # hidden layer; take any multiple of 8 as an argument in Dense()
model.add(tf.layers.Dense(10, activation = 'softmax'))
```

```python
# Compile the model
model.compile(loss = 'sparse_categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
# Scale the data
x_train = x_train / 255
x_test = x_test / 255
x_train[0]
```

```python
# Training the model
model.fit(x_train, y_train, epochs = 15)
```

```python
y_pred = model.predict(x_test)
```

```python
model.evaluate(x_test, y_test)
```
